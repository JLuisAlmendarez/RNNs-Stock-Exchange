```python
# ------------------------------
# Script completo: AAPL + Benchmark + Indicadores técnicos
# ------------------------------

import yfinance as yf
import pandas as pd
import talib
import numpy as np

# --- Descarga de datos ---
tickers = ['AAPL', 'SPY', '^VIX', 'XLK']
data = {t: yf.download(t, start='2018-01-01', end='2025-01-01') for t in tickers}

# --- Datos de AAPL ---
# Seleccionamos solo columnas planas para evitar MultiIndex
aapl = data['AAPL'][['Open','High','Low','Close','Volume']].copy()

# --- Features Técnicos ---
# Medias móviles
aapl['SMA_20'] = aapl['Close'].rolling(window=20).mean()
aapl['SMA_50'] = aapl['Close'].rolling(window=50).mean()

# Volatilidad rolling
aapl['Vol_20'] = aapl['Close'].rolling(window=20).std()

# RSI 14
aapl['RSI_14'] = talib.RSI(aapl['Close'].to_numpy().ravel(), timeperiod=14)

# MACD
macd, macd_signal, macd_hist = talib.MACD(aapl['Close'].to_numpy().ravel(),
                                          fastperiod=12, slowperiod=26, signalperiod=9)
aapl['MACD'] = macd
aapl['MACD_signal'] = macd_signal
aapl['MACD_hist'] = macd_hist

# Bollinger Bands
bb_upper, bb_middle, bb_lower = talib.BBANDS(aapl['Close'].to_numpy().ravel(),
                                             timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
aapl['BB_upper'] = bb_upper
aapl['BB_middle'] = bb_middle
aapl['BB_lower'] = bb_lower

# --- Integrar SPY, VIX y XLK ---
# Seleccionar solo Close y renombrar
spy_close = data['SPY']['Close'].copy()
spy_close.name = "SPY_Close"

vix_close = data['^VIX']['Close'].copy()
vix_close.name = "VIX_Close"

xlk_close = data['XLK']['Close'].copy()
xlk_close.name = "XLK_Close"

# Merge con índice de fechas
aapl = aapl.merge(spy_close, left_index=True, right_index=True, how='left')
aapl = aapl.merge(vix_close, left_index=True, right_index=True, how='left')
aapl = aapl.merge(xlk_close, left_index=True, right_index=True, how='left')

# --- Guardar a CSV ---
aapl.to_csv('AAPL_full_dataset.csv')

print("✅ Dataset generado con éxito: AAPL + SPY + VIX + XLK + indicadores técnicos.")
```
